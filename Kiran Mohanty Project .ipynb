#### Q2_Dataset
#### Create spend summary at Brand, Channel &amp; Promo level
#### Aggregate spends to monthly level
#### Split Spend based on number of days in Start Date &amp; End Date.

```python
Q2_Dataset.columns=["Brand","Channel","Start Date","End Date","Promo","Spend"]
```

```python
Q2_Dataset
```

```python
# (i)Create spend summary at Brand, Channel & Promo level
summary = Q2_Dataset.groupby(['Brand','Channel','Promo'])['Spend'].sum()
# (ii)Aggregate spends to monthly level
Q2_Dataset['Start Date'] = pd.to_datetime(Q2_Dataset['Start Date'])
Q2_Dataset['End Date'] = pd.to_datetime(Q2_Dataset['End Date'])
Q2_Dataset['Month'] = Q2_Dataset['Start Date'].dt.to_period('M')
monthly_summary = Q2_Dataset.groupby(['Brand', 'Channel', 'Promo', 'Month'])['Spend'].sum()

# (iii)Split Spend based on number of days in Start Date & End Date
Q2_Dataset['Duration'] = (Q2_Dataset['End Date'] - Q2_Dataset['Start Date']).dt.days + 1
Q2_Dataset['Daily Spend'] = Q2_Dataset['Spend'] / Q2_Dataset['Duration']

split_spend = Q2_Dataset[['Brand', 'Channel', 'Promo', 'Start Date', 'End Date', 'Daily Spend']]

# Print the results
print("Spend summary at Brand, Channel & Promo level:")
print(summary)
print()

print("Monthly spend summary at Brand, Channel & Promo level:")
print(monthly_summary)
print()

print("Split Spend based on number of days in Start Date & End Date:")
print(split_spend)
```

```python
# Combine “sales” &amp; “iris” data to make the base dataframe
```

```python
# Create DataFrames from the sample data
sales_df = pd.DataFrame(Sales)
iris_df = pd.DataFrame(Q3_iris)

# Set the common columns 'product_id', 'geography_id', and 'period' as the index for both DataFrames
sales_df.set_index(['product_id', 'geography_id', 'period'], inplace=True)
iris_df.set_index(['product_id', 'geography_id', 'period'], inplace=True)

# Merge the DataFrames based on their index
base_df = pd.merge(sales_df, iris_df, left_index=True, right_index=True)

base_df
```

```python
base_df.describe()
```

```python
#Define a function to convert '2015 Spend' data to weekly data and distribute proportionally

def convert_to_weekly_spend(row):
    # Get the total volume for the product_id and geography_id
    total_volume = row['volume']
    # Calculate the weekly spend based on 'adwofpa', 'afdpa', 'afwodpa' values
    weekly_spend = row[['adwofpa', 'afdpa', 'afwodpa']] * total_volume
    return pd.Series(weekly_spend.values, index=['weekly_adwofpa', 'weekly_afdpa', 'weekly_afwodpa'])

# Calculate the weekly spend data as a separate DataFrame
weekly_spend_data = base_df.apply(convert_to_weekly_spend, axis=1)

# Concatenate the weekly spend data to the base DataFrame along the columns
base_df = pd.concat([base_df, weekly_spend_data], axis=1)

# Weekly spend for each 'parent_id' sums back to the original value
weekly_spend_sum = base_df.groupby(['product_id', 'geography_id', 'period'])[['weekly_adwofpa', 'weekly_afdpa', 'weekly_afwodpa']].sum()
base_df[['weekly_adwofpa', 'weekly_afdpa', 'weekly_afwodpa']] = weekly_spend_sum

# Reset the index for a clean output (optional)
base_df.reset_index(inplace=True)
```

```python
# Display the final base DataFrame
print(base_df)
```
