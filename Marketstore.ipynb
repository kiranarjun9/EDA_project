#### 1.What is the correlation between discounts and the quantity of products sold?

```python
def histogram_intersection(discounts,quantity):
    k = np.minimum(discounts,quantity).sum().round(decimals=1)
    return k
store.corr(method=histogram_intersection)
```

```python
store.corr()
```

```python
sns.heatmap(store.corr(method='pearson'),cmap='PiYG',annot = True)     #Correlation techinques:-'pearson', 'spearman', 'kendall',
```

### 6.What are the outliers in an Profits?

```python
Q1=store.Profit.quantile(0.25)
Q2=store.Profit.quantile(0.50)
Q3=store.Profit.quantile(0.75)
print("First Quratile:",Q1)
print("Second Quartile:",Q2)
print("Third Quartile:",Q3)
```

```python
IQR=Q3-Q1
print("Inter Quartile Range:",IQR)
```

```python
HE=Q3+1.5*IQR
print("HIGHER END POINT:",HE)

LE=Q1-1.5*IQR
print("LOWER END POINT:",LE)
```
